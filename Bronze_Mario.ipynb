Imports

```python
import os
import pandas as pd 
import datetime
```

Setting path variables

```python
cwd =os.getcwd()
print(f"File Path:{cwd}")

folder= "\\mario_data"

data_path=cwd+folder
print(f"Data Location: {data_path}")
```

Get all folders and files

```python
folder_files= os.listdir(data_path)
for f in folder_files:
    print(f)
```

Taking a peek of the data

```python
df_peek= pd.read_csv(data_path + '/mario_data_20250101.csv')

df_peek.head()
```

Combining all csvs and json files

```python
df_list=[]

for file in folder_files:
    print(f"Reading file: {file}")
    file_path= data_path+'\\'+file

    if file.endswith('jsonl'):
        temp_df= pd.read_json(file_path)
    else:
        temp_df=pd.read_csv(file_path)
    temp_df['filename']=file
    temp_df['loadDateTimeStamp']=datetime.datetime.now()
    
    df_list.append(temp_df)

df= pd.concat(df_list,ignore_index=True)
print('Completed Dataframe Concatenation')
```

Saving Bronze Layer

```python
df.to_parquet('./Bronze layer/Bronze.parquet',index=False)
```
