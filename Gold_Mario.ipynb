Total team points, Average Rank, Total coins spent in Toad Town for each team

```python
rank_map = {'A': 4, 'B': 3, 'C': 2, 'D': 1}
# Making new columns which correspond to the numeric values of the each letter rank
for col in ['Kart Racing Rank', 'Platforming Rank', 'Boss Battle Rank']:
    df[col + ' Num'] = df[col].map(rank_map)

# Compute average rank for each team by avergaing all three ranks
df['Average Rank'] = df[['Kart Racing Rank Num', 'Platforming Rank Num', 'Boss Battle Rank Num']].mean(axis=1)
# grouping team points corresponding to each team
total_team_points= df.groupby('Team')['Team Points'].sum().reset_index().rename(columns={'Team Points': 'Total Team Points'})
# grouping average rank for each team
team_avg_rank = df.groupby('Team')['Average Rank'].mean().reset_index().rename(columns={'Average Rank': 'Average Rank Score'})

total_coins_spent = df.groupby('Team')['Coins Spent in Toad Town'].sum().reset_index().rename({'Coins Spent in Toad Town':'Total Coins spent in Toad Town '})
print(f"{total_team_points}\n")
print(f"{team_avg_rank}\n")
print(total_coins_spent)
```

```python
# aggregating all of the three columns into one dataframe
gold_report1=df.groupby('Team', as_index=False).agg({'Team Points': 'sum','Average Rank': 'mean','Coins Spent in Toad Town': 'sum'}).rename(columns={
          'Team Points': 'Total Team Points',
          'Average Rank': 'Average Rank Score',
          'Coins Spent in Toad Town': 'Total Coins Spent in Toad Town' })
```

```python
gold_report1
```

Saving results to parquet file

```python
gold_report1.to_parquet('./Gold layer/gold_team_summary.parquet')
```

Total Team Points earned by each player sorted in descending

```python
gold_report2 = df.groupby('Player Name')['Team Points'].sum().reset_index().sort_values('Team Points',ascending=False,ignore_index=True).rename(columns={'Team Points': 'Total Team Points earned'})
```

```python
gold_report2
```

```python
gold_report2.to_parquet('./Gold layer/gold_player_performance.parquet')
```

Count and Average Number of Power-ups used per player

```python
gold_report3= df.groupby('Player Name',as_index=False).agg(Power_Ups=('Power-Ups Used','count'),Avg_PowerUps=('Power-Ups Used','mean')).sort_values(by='Power_Ups',ignore_index=True)
```

```python
gold_report3
```

```python
gold_report3.to_parquet('./Gold layer/gold_powerup_player.parquet')
```

Average Number of Lives Lost in each world in descending order

```python
gold_report4= df.groupby('World')['Lives Lost'].mean().reset_index().rename(columns={'Lives Lost':'Average Lives Lost'}).sort_values('Average Lives Lost',ascending=False)
```

```python
gold_report4
```

```python
gold_report4.to_parquet('./Gold layer/gold_world_difficulty.parquet')
```

Top Player per Team by Team Points

```python

gold_report5=df.loc[df.groupby('Team')['Team Points'].idxmax(), ['Team', 'Player Name', 'Team Points']].reset_index(drop=True).sort_values('Team Points', ascending=False)

```

```python
gold_report5.rename(columns={'Player Name':'Top Player'})
```

```python
gold_report5.to_parquet('./Gold layer/gold_top_player_per_team.parquet')
```

Popularity of each vehicle

```python
gold_report6=df.loc[:,'Vehicle Type'].value_counts().reset_index()
```

```python
gold_report6.rename(columns={'index':'Vehicle Type','count':'Usage Count'})
```

```python
gold_report6.to_parquet('./Gold layer/gold_vehicle_counts.parquet')
```

Players who are at highest risk( by lives lost)

```python
gold_report7 = df.groupby('Player Name')['Lives Lost'].sum().reset_index().sort_values('Lives Lost',ignore_index=True,ascending=False).rename(columns={'Lives Lost':'Total Lives Lost'})
```

```python
gold_report7
```

```python
gold_report7.to_parquet('./Gold layer/gold_risk_assessment.parquet')
```

Average and Total number of levels completed in each world

```python
gold_report8=df.groupby('World',as_index=False).agg(Number_of_Levels_Completed=('Levels Completed','sum'), Average_Levels=('Levels Completed','mean')).rename(columns={'index':'World','Number_of_Levels_Completed':'Total Number of levels completed','Average_Levels':'Average Levels Completed'})
```

```python
gold_report8
```

```python
gold_report8.to_parquet('./Gold layer/gold_world_completion.parquet')
```

Total and Average number of times each team was hit by enemies

```python
gold_report9=df.groupby('Team',as_index=False).agg(Total=('Times Hit by Enemies','sum'),Average=('Times Hit by Enemies','mean')).rename(columns={'Total':'Total Number of Times Hit by Enemies','Average':'Average Number of times Hit by Enemies'})
```

```python
gold_report9
```

```python
gold_report9.to_parquet('./Gold layer/gold_hits_team.parquet')
```
